```python
# for dataframe operations
import pandas as pd

# data visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lets import the dataset using the read_csv function
data = pd.read_csv('LoanData.csv')

# lets check the shape of the dataset
data.shape
```

### Make a Heat Map

```python
plt.rcParams['figure.figsize'] = (5, 5)
plt.style.use('fivethirtyeight')

sns.heatmap(data.corr(), 
            annot = True, 
            fmt = '0.01',
            cmap = 'Wistia')
plt.title('Correlation Heatmap')
plt.show()
```

### Compare Loan Amount and Applicant Income

```python
sns.scatterplot(data['ApplicantIncome'], data['LoanAmount'])
plt.show()
```

### Find the Number of applicants where the Co applicant's Income Higher

```python
data[data['ApplicantIncome'] < data['CoapplicantIncome']][['ApplicantIncome',
                                'CoapplicantIncome']].sort_values(by = 'CoapplicantIncome', ascending = False)
```
